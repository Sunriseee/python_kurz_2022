```python
import pandas
import matplotlib.pyplot as plt

import requests

r = requests.get(
    "https://raw.githubusercontent.com/andywaltlova/python-1-podzim-2022/master/ukoly/data/platy_2021_02.csv")
open("platy_2021_02.csv", "wb").write(r.content)

platy = pandas.read_csv("platy_2021_02.csv")
platy
```

```python
import requests

r = requests.get(
    "https://raw.githubusercontent.com/andywaltlova/python-1-podzim-2022/master/ukoly/data/temperature.csv")
open("temperature.csv", "wb").write(r.content)
teplota = pandas.read_csv("temperature.csv")
teplota
```

```python

teplota = teplota.set_index(["City", "Day"])
teplota
```

```python


vybrana_mesta = (teplota.loc[["Helsinki","Tokyo","Miami Beach"]])
vybrana_mesta = vybrana_mesta.notna()
```

```python

import seaborn as sns
import matplotlib.pyplot as plt


sns.boxplot(x=["City"], y=["AvgTemperature"], data = vybrana_mesta)
plt.show()
```
