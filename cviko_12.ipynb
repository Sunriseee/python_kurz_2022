```python
import pandas
import matplotlib.pyplot as plt

url = "https://kodim.cz/cms/assets/kurzy/python-data-1/python-pro-data-1/vizualizace/ucet.csv"
ucet = pandas.read_csv(url, names=['datum', 'pohyb'], index_col='datum')

ucet
```

```python
ucet.plot() # zobraz graf!
plt.show     #zobraz data 
```

```python
ucet.cumsum().plot()  # kumulativn suma
plt.show
```

```python
ucet.plot(
    kind='bar',
    color='yellow',
    grid=True
)

plt.show()
```

```python
color = ["red" if value < 0 else "green" for value in ucet["pohyb"]]

ucet["pohyb"].plot(kind='bar',
                   color=color,
                   width=0.5,

                   # Popisky
                   ylabel="stka v K",
                   xlabel="Datum transakce",
                   title="Transakce na t"
                   )

plt.show()
```

```python
# fig - figure
# axes - ax

color = ["#EB5160" if value < 0 else "#91C499" for value in ucet["pohyb"]]

fig, ax = plt.subplots(figsize=(15, 7))
print(type(fig))
print(type(ax))


ax.set_title("Pohyby na tu", fontweight="bold", fontsize=24)
ax.tick_params(axis="x", labelrotation=45)
ax.set_ylabel("stka v K", color="darkgrey", fontsize=12)
ax.bar(ucet.index, ucet.pohyb, color=color)
plt.axhline(y=0, color="k", linestyle="--")

# Popisky hodnot v grafu
for i in range(len(ucet.pohyb)):
    plt.annotate(str(ucet.pohyb[i]),
                 xy=(ucet.index[i],
                     ucet.pohyb[i]),
                 ha='center',
                 va='bottom')

ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

plt.show

#  https: // matplotlib.org/3.1.1/api/_as_gen/matplotlib.axes.Axes.set_title.html
```
